```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 基本参数
# =========================

g_moon = 1.62   # 月球重力 (m/s^2)
g_earth = 9.8   # 地球重力 (m/s^2)

v0 = 5.0        # 初始速度 (m/s)
dt = 0.01       # 时间步长


# =========================
# 模拟函数
# =========================
def simulate_jump(g):
    y = 0
    v = v0
    trajectory = []

    for _ in range(2000):
        v = v - g * dt
        y = y + v * dt

        if y < 0:
            break

        trajectory.append(y)

    return trajectory


# =========================
# 运行模拟
# =========================
moon_traj = simulate_jump(g_moon)
earth_traj = simulate_jump(g_earth)


# =========================
# 画图
# =========================
plt.figure()

plt.plot(moon_traj, label="Moon (g = 1.62)")
plt.plot(earth_traj, label="Earth (g = 9.8)")

plt.title("Jump Comparison: Moon vs Earth")
plt.xlabel("Time Step")
plt.ylabel("Height")
plt.legend()

# 保存图片（用于GitHub展示）
plt.savefig("comparison.png")

plt.show()


# =========================
# 输出数据（加分项）
# =========================
print("Moon max height:", max(moon_traj))
print("Earth max height:", max(earth_traj))
```
